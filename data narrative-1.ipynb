```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
var= pd.read_csv("C:\\Users\\vatsal trivedi\\Documents\\Data narrative assign-1\\books.csv")
plt.figure(figsize=(10,5))
bn = var.groupby('language_code')['title'].count().plot.bar()
plt.title('No. of books v/s language')
plt.xlabel('Language')
plt.xticks(fontsize = 11)
plt.ylabel('Number of Books')
for q in bn.patches:
    bn.annotate(str(q.get_height()), (q.get_x()-0.4, q.get_height()+110))
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
var= pd.read_csv("C:\\Users\\vatsal trivedi\\Documents\\Data narrative assign-1\\books.csv")
most_reviews = var.sort_values('work_text_reviews_count', ascending = False).head(10).set_index('title')
plt.figure(figsize=(10,7))
sns.barplot(x = 'work_text_reviews_count', y =most_reviews.index , data = most_reviews, palette='Greens_d')
plt.title("Top 10 Books with most reviews", fontsize=16)
plt.xlabel("Number of reviews", fontsize=14)
plt.ylabel("Title of Book", fontsize=14)
```
