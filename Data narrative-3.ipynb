```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
var= pd.read_csv("C:\\Users\\vatsal trivedi\\Documents\\data narr-3\\usopen men.csv")
var1=pd.read_csv("C:\\Users\\vatsal trivedi\\Documents\\data narr-3\\ausopen men.csv")
var3= pd.read_csv("C:\\Users\\vatsal trivedi\\Documents\\data narr-3\\french open men.csv")
var4= pd.read_csv("C:\\Users\\vatsal trivedi\\Documents\\data narr-3\\wimbledon men.csv")
```

```python
X = var[['FSW.1', 'FSP.1']].copy()
n_clusters = 5
gmm_model = GaussianMixture(n_components=n_clusters)
gmm_model.fit(X)
cluster_labels = gmm_model.predict(X)
X = pd.DataFrame(X)
X['cluster'] = cluster_labels

for k in range(0,n_clusters):
    data = X[X["cluster"]==k]
    plt.scatter(data["FSP.1"],data["FSW.1"])
plt.title("Clusters by GaussianMixture of usopen-men")    
plt.ylabel("Number of first serves won by player-1")
plt.xlabel("First serve % for player-1")
plt.show()

X = var[['FSW.2', 'FSP.2']].copy()
n_clusters = 5
gmm_model = GaussianMixture(n_components=n_clusters)
gmm_model.fit(X)
cluster_labels = gmm_model.predict(X)
X = pd.DataFrame(X)
X['cluster'] = cluster_labels

for k in range(0,n_clusters):
    data = X[X["cluster"]==k]
    plt.scatter(data["FSP.2"],data["FSW.2"])
plt.title("Clusters by GaussianMixture of usopen-men")    
plt.ylabel("Number of first serves won by player-2")
plt.xlabel("First serve % for player-2")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X = var8['NPA.1']
y= var8['NPW.1']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=100)
X_train = X_train[:,np.newaxis]
X_test = X_test[:,np.newaxis]
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
c = [i for i in range (1,len(y_test)+1,1)]
plt.plot(c,y_test,color='r',linestyle='-',label='Actual')
plt.plot(c,y_pred,color='b',linestyle='-',label='Predicted')
plt.xlabel('Net Points Attempted by Player-1')
plt.ylabel('Net Points Won by Player-1')
plt.title('Prediction')
plt.legend()
plt.show()

X = var8['BPC.2']
y= var8['BPW.2']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=100)
X_train = X_train[:,np.newaxis]
X_test = X_test[:,np.newaxis]
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
c = [i for i in range (1,len(y_test)+1,1)]
plt.plot(c,y_test,color='r',linestyle='-',label='Actual')
plt.plot(c,y_pred,color='b',linestyle='-',label='Predicted')
plt.xlabel('Break Points Created by Player-2')
plt.ylabel('Break Points Won by Player-2')
plt.title('Prediction')
plt.legend()
plt.show()
```
